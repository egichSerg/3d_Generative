### Нормализуем

```python
center = verts.mean(0)
verts = verts - center
scale = verts.max()
verts /= scale
verts = verts.unsqueeze(0)
verts.shape
```

### создаем исходный объект

```python
src_mesh = ico_sphere(4, device)
```

```python
src_vert = src_mesh.verts_list()
deform_verts = torch.full(src_vert[0].shape, 0.0, device=device, requires_grad=True)
```

```python
optimizer = torch.optim.SGD([deform_verts], lr = 1.0, momentum = 0.9)
```

```python
w_chamfer = 1.0
w_edge = 1.0
w_normal = 0.01
w_laplacian = 0.1
```

```python
epochs = 2000
for epoch in range(epochs):
    
    print(f"Epoch {epoch}")

    # fwd pass
    new_src_mesh = src_mesh.offset_verts(deform_verts)
    sample_trg = verts
    sample_src = sample_points_from_meshes(new_src_mesh,
    verts.shape[1])
    
    # calc loss
    loss_chamfer, _ = chamfer_distance(sample_trg, sample_src)
    loss_edge = mesh_edge_loss(new_src_mesh)
    loss_normal = mesh_normal_consistency(new_src_mesh)
    loss_laplacian = mesh_laplacian_smoothing(new_src_mesh, method='uniform')
    loss = (
    loss_chamfer * w_chamfer
    + loss_edge * w_edge
    + loss_normal * w_normal
    + loss_laplacian * w_laplacian
    )

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

### сохраняем гениальную полученную модель

```python
final_verts, final_faces = new_src_mesh.get_mesh_verts_faces(0)
final_verts = final_verts * scale + center

final_obj = os.path.join("./", "deform1.ply")
save_ply(final_obj, final_verts, final_faces, ascii=True)
```
